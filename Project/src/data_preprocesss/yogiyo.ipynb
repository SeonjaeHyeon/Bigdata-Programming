```python
import json
import pandas as pd
import csv

import os
import os.path
```

```python
dir_path = 'yogiyo'
work_dir = os.getcwd()
```

```python
project_dir = os.path.join(work_dir, dir_path)
project_dir
```

```python
LOCATIONS = [
    "강남구", "강동구", "강북구", "강서구", "관악구", "광진구", "구로구", "금천구", "노원구", "도봉구", "동대문구", "동작구", "마포구",
    "서대문구", "서초구", "성동구", "성북구", "송파구", "양천구", "영등포구", "용산구", "은평구", "종로구", "중구", "중랑구"
]
```

```python
dir_list = os.listdir(project_dir)
dir_list
```

```python
file_list = []

for directory in dir_list:
    for (path, _, files) in os.walk(os.path.join(project_dir, directory)):
        file_list.append({'path': path, 'files': files})

file_list
```

```python
data = {location: None for location in LOCATIONS}
data
```

```python
for location in LOCATIONS:
    df_data = None
    for directory in dir_list:
        json_file = open(os.path.join(project_dir, directory, f'rank-{location}.json'), 'r', encoding='utf-8')
        json_data = json.load(json_file)
        temp = pd.DataFrame(json_data['restaurants'])
        temp['date'] = directory
        temp['rank'] = range(1, len(temp) + 1)

        if df_data is None:
            df_data = temp
        else:
            df_data = pd.concat([df_data, temp], axis=0)

    data[location] = df_data[['id', 'name', 'review_avg', 'rank']]
    # data[location] = df_data

data['강남구']
```

```python
for location in data:
    data[location].to_csv(f'yogiyo-{location}.csv', encoding='utf-8-sig', quoting=csv.QUOTE_ALL)
```
